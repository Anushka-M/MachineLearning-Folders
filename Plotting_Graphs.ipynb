```python
#plot function x^3
import numpy as np
x = np.arange(1,10,2)
y = x**2
plt.plot(x,y,"g--")
plt.show()
```

```python
#plot function x^3
import numpy as np
x = np.arange(1,10,0.5)
y = x**2
y2 = x**0.5
plt.axis([0,10,0,100])
plt.plot(x,y,"blue",marker="p",linewidth=0.5,label="x^2")
plt.plot(x,y2,"red",marker="p",linewidth=0.5,label="x^0.5")
plt.legend()
plt.xlabel("age")
plt.grid()
plt.show()
```
